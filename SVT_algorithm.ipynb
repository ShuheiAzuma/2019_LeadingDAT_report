### 内容1 低ランク化ステップの実装

#### 入力：欠損したn1 X n2の行列M，m個の欠損していない要素を示すΩ(・)

```python
M = np.loadtxt('./kadai/kadai/sea_missing.csv', delimiter = ",")
Omega = np.loadtxt('./kadai/kadai/sea_mask.csv', delimiter = ",")
```

##### 行数n1, 列数n2の取得

```python
n1 = M.shape[0]
n2 = M.shape[1]
```

##### m:欠損していない要素数

```python
m = Omega.sum()
```

#### Y0を適当な行列とする．；(Mと同じ大きさの要素はすべて0の行列)

```python
Y0 = np.zeros((n1, n2))
```

#### パラメータの設定．；τとσ

```python
tau = 2.5 *(n1 + n2)
delta = 1.2 * n1 * n2 /m 
```

#### Singular Value Decompotion; Yを特異値分解して三つの行列U, S, Vtを得る．

```python
u, s, vt = np.linalg.svd(Y0, full_matrices=True)
```

#### Sの全要素からτをひき，負の要素をゼロとした行列をS_tildeとする．

```python
s_tilde = s - tau
index_minus = np.where(s_tilde < 0)
s_tilde[index_minus] = 0
s_tilde_mat = np.zeros((n1, n2))
s_tilde_mat[:n1, :n1] = np.diag(s_tilde)
```

#### S_tildeを使って低ランク行列Xを作る．

```python
X = np.dot(np.dot(u, s_tilde_mat), vt)
```

### 内容2 誤差減少化ステップの実装

```python
Y = np.zeros((n1, n2))
Y = Y + delta * Omega * (M - X)
```

### 内容3 SVTアルゴリズムの実装

```python
M = np.loadtxt('./kadai/kadai/sea_missing.csv', delimiter = ",")
Omega = np.loadtxt('./kadai/kadai/sea_mask.csv', delimiter = ",")
n1, n2 = M.shape
m = Omega.sum()
Y0 = np.zeros((n1, n2))

Y = Y0
k_max = 1000
```

```python
def svt_algorithm(Y):
    
    # Set previous Y
    Y = Y
    
    # define parameter
    tau = 2.5 *(n1 + n2)
    delta = 1.2 * n1 * n2 /m 
    
    # Singular Value Decompotion
    u, s, vt = np.linalg.svd(Y)
    
    # Amend "s" to Matrix(named s_tilde_mat)
    s_tilde = s - tau
    index_minus = np.where(s_tilde < 0)
    s_tilde[index_minus] = 0
    s_tilde_mat = np.zeros((n1, n2))
    s_tilde_mat[:n1, :n1] = np.diag(s_tilde)
    
    # Update X
    X = np.dot(np.dot(u, s_tilde_mat), vt)
    
    # Update Y
    Y = Y + delta * Omega * (M - X)
    
    return(X, Y)
```

```python
for i in range(1, k_max):
    
    X, Y = svt_algorithm(Y)
    
    Y = Y
```

```python
M = np.loadtxt('./kadai_option/kadai_option/cor_missing.csv', delimiter = ",")
Omega = np.loadtxt('./kadai_option/kadai_option/cor_mask.csv', delimiter = ",")
n1, n2 = M.shape
m = Omega.sum()
Y0 = np.zeros((n1, n2))

Y = Y0
k_max = 1000
```

```python
def svt_algorithm_option(Y):

    # Set previous Y
    Y = Y
    
    # define paramete
    tau = 100 *(n1 + n2)
    delta = 0.8 * n1 * n2 /m 
    
    # Singular Value Decompotion
    u, s, vt = np.linalg.svd(Y)
    
    # Amenndment "s" to Matrix(named s_tilde_mat)
    s_tilde = s - tau
    index_minus = np.where(s_tilde < 0)
    s_tilde[index_minus] = 0
    s_tilde_mat = np.zeros((n1, n2))
    s_tilde_mat[:n1, :n1] = np.diag(s_tilde)
    
    # Update X
    X = np.dot(np.dot(u, s_tilde_mat), vt)
    
    # Update Y
    Y = Y + delta * Omega * (M - X)
    
    return(X, Y)
```

```python
for i in range(1, k_max):
    
    X, Y = svt_algorithm_option(Y)
    
    Y = Y
```
